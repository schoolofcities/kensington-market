```python
import pandas as pd
pd.set_option('display.max_columns', None)

cleaned_df = pd.read_csv("2025-cleaned.csv")
messy_df = pd.read_csv("all-messy.csv")
```

```python
messy_df = messy_df[['primary_address', 'name_2025', 'category_2025', 'sub_category_2025', 'long', 'lat']]
messy_df.head()
```

```python
joined_df = cleaned_df.merge(messy_df, left_on='address', right_on='primary_address', how='outer')
joined_df.head(50)
```
